```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Normal Distribution

```python
x=np.random.normal(loc=5,scale=3,size=10000)
mean=np.mean(x)
```

```python
ax=sns.histplot(x,bins=50,kde=True,color="#42EE26")
ax.axvline(mean, 0,0.95,color="red",linewidth=4,label="agdfhdhdhahdg")
ax.set(title='Mean')
ax.set_facecolor((1.0, 0.47, 0.42))
plt.show()
```

```python
heights=np.random.normal(loc=160,scale=10,size=10000)
```

```python
height_mean=np.mean(heights)
```

```python
ax=sns.histplot(heights,bins=50,kde=True,color="green")
ax.axvline(height_mean, 0,0.95,color="red",linewidth=4,label="agdfhdhdhahdg")
ax.set(title='Mean')
ax.set_xlabel("height of people")
plt.show()
```
